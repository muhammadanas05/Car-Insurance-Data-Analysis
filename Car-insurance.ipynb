```python
import pandas as pd
import numpy as np
from statsmodels.formula.api import logit

df = pd.read_csv('car_insurance.csv')
df.head()
```

```python
df = df.drop(columns=['id'])

best_feature = None
best_accuracy = 0
```

```python
for feature in df.columns:
    if feature == 'outcome':
        continue

    formula = f'outcome ~ {feature}'
    model = logit(formula, data=df).fit(disp=False)
    
    predictions = model.predict(df) > 0.5
    
    accuracy = (predictions == df['outcome']).mean()

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_feature = feature
```

```python
best_feature_df = pd.DataFrame({'best_feature': [best_feature], 'best_accuracy': [best_accuracy]})

best_feature_df
```
